```python
import nltk
from nltk.stem.porter import PorterStemmer

```

```python
stemmer=PorterStemmer()
```

```python
def token(txt):
    tokens = nltk.word_tokenize(txt)
    a = [stemmer.stem(w) for w in tokens]
    return " ".join(a)

# Test the function #example
print(token("you are beautiful, beauty"))
```

```python
df['text'].apply(lambda x: token(x))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
tfid= TfidfVectorizer(analyzer='word', stop_words='english')
```

```python
tfid
```

```python
matrix=tfid.fit_transform(df['text'])
```

```python
matrix
```

```python
similer =cosine_similarity(matrix)
```

```python
similer
```

Recommender Function

```python
def recommender(song_name):
    if song_name in df['song'].values:
        idx = df[df['song'] == song_name].index[0]
        distance = sorted(list(enumerate(similer[idx])), reverse=True, key=lambda x: x[1])
        song = []
        for s_id in distance[1:7]:
            song.append(df.iloc[s_id[0]].song)
        return song
    else:
        return "Song not found in the database"
        
```

```python
recommender('Wisdom Justice And Love')
```
